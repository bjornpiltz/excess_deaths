```r
library(tidyverse)

options(repr.plot.width = 7, repr.plot.height = 8, repr.plot.res = 240,
        readr.show_col_types = F)

State = "Texas"

read_rds("https://github.com/bjornpiltz/excess_deaths/raw/main/data/us_excess_deaths_per_age_group.rds")%>%
  filter(Jurisdiction == State)%>%
  ggplot(aes(x = Date, 
             y = `Number of Deaths`/Population*1e6, 
             linetype = Type)) +
  labs(x = "", y = "", title = paste("Weekly deaths per million in", State, "Compared to he baseline")) + 
  geom_line() +
  facet_wrap(~`Age Group`, scales = "free_y", ncol = 1) +
  theme(legend.position = "top")
```
